```python
def place():
  place_list = ['부엌', '작은방', '안방', '추리의 방']
  place_dict = {'부엌':kitchen, '작은방':smallroom, '안방':bedroom, '추리의 방':detect1}
  print(place_list)
  print('추리의 방은 사건의 전말을 추리할 수 있는 방입니다.')
  print()
  while True:
    a = input('어디를 가겠습니까? ')
    print()
    if a in place_list:
      return place_dict[a]()
```

```python
def detect1():
  time.sleep(1)
  print(f'사건의 전말을 추리해주세요.')
  time.sleep(1)
  mur = input('Mrs.John이 그의 남편을 죽인 것이 확실합니까? (yes or no): ')
  if mur == '1':
    print('맞습니다. 그녀가 남편을 살해했습니다.')
    return detect2()
  else:
    print('틀렸습니다. 다시 추리해야 합니다.')
    return replay()
```

```python
def detect2():
  tool_list = ['로프', '총', '칼']
  time.sleep(1)
  print(tool_list)
  tool = input('그녀는 남편을 무엇으로 살해했습니까? (위의 보기에서 고르시오) : ')
  if tool == '로프':
    print('맞습니다. 로프가 살해 도구입니다.')
    return detect3()
  else:
    print('틀렸습니다. 다시 추리해야 합니다.')
    return replay()
```

```python
def detect3():
  motive_list = ['쨈', '카나리아', '퀼트', '앞치마']
  time.sleep(1)
  print(motive_list)
  motive = input('살해 동기를 알아낼 수 있는 결정적 단서는 무엇입니까? (위의 보기에서 고르시오) : ')
  if motive == '카나리아' :
    print('맞습니다. 카나리아 새가 결정적인 살해 동기입니다.')
    return allthing()
  else:
    print('틀렸습니다. 다시 추리해야 합니다.')
    return replay()
```

```python
def replay():
  while True :
    response = input('다시 하시겠습니까?(yes or no): ')
    print()
    if response == '1' :
      return place()
    elif response == '2' :
      while True:
        wannaknow = input('사건의 전말을 알아보겠습니까? (yes or no): ')
        print()
        if wannaknow == '1' :
          return allthing()
        elif wannaknow == '2' :
          return print('감사합니다!')
```
